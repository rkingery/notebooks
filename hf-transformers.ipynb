````python
import torch
from transformers import *
````

````python
device = 'mps'
````

## Sentiment Analysis

````python
model = pipeline('text-classification', model='distilbert-base-uncased-finetuned-sst-2-english', device=device)
````

````python
model.model
````

````python
sum([p.numel() for p in model.model.parameters()])
````

````python
text = 'I was not necessarily super pleased with this restaurant, and may avoid returning in the future.'
pred = model(text)
pred
````

````python
del model
````

## Text Generation

````python
model = pipeline('text-generation', model='openai-community/gpt2', device=device)
````

````python
model.model
````

````python
sum([p.numel() for p in model.model.parameters()])
````

````python
text = 'When I was younger, I remember'
pred = model(text, max_length=100)
print(pred[0]['generated_text'])
````

````python
del model
````

## Text Summarization

````python
model = pipeline('summarization', model='facebook/bart-large-cnn', device=device)
````

````python
model.model
````

````python
sum([p.numel() for p in model.model.parameters()])
````

````python
text = """
Presidential elections were held in Russia from 15 to 17 March 2024. It was the eighth presidential election in the country. 
The incumbent president Vladimir Putin won with 88% of the vote, the highest percentage in a presidential election in post-Soviet Russia, 
gaining a fifth term in what was widely viewed as a foregone conclusion. He is scheduled to be inaugurated on 7 May 2024.
"""
pred = model(text, max_length=60)
print(pred[0]['summary_text'])
````

````python
del model
````

## Question Answering

````python
model = pipeline('question-answering', model='distilbert/distilbert-base-cased-distilled-squad', device=device)
````

````python
model.model
````

````python
sum([p.numel() for p in model.model.parameters()])
````

````python
context = 'John lives with his wife and three children in a large wooden house.'
question = 'How many children does John have?'
pred = model(context=context, question=question)
print(pred['answer'])
````
